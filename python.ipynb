### 문자열 "My name is seunga."에 대해 다음 물음에 답하라.
- 문자열의 문자수를 출력하라.
- 문자열의 10번 반복한 문자열을 출력하라.
- 문자열의 첫 번째 문자를 출력하라.
- 문자열에서 처음 4문자를 출력하라.
- 문자열의 문자를 거꾸로 출력하라.

### 다음과 같은 게임 프로그램을 작성하라.
- 플레이어가 처음에 50 달러를 가지고 있다. 동전을 한 번 던져서 앞면(1) 또는 뒷면(2)이 나온다. 맞추면 9 달러를 따고 틀리면 10 달러를 잃는다. 플레이어가 돈을 모두 잃거나 100달러가 되면 게임이 종료된다.

```python
from random import randint

money = 50

while money > 0 and money < 100:
    coin = randint(1,2)
    if coin == 1:
        money += 9
    elif coin == 2:
        money -= 10
    print(coin, money)
```

### 두 수의 최대 공약수는 두 수를 나누어 떨어지는 가장 큰 수이다. 
### 예를 들어 (16, 24)의 최대 공약수는 8이다. 
### 두 수를 입력받아 다음 알고리즘에 의해 최대 공약수를 구하는 프로그램을 작성하라.
- 큰 수를 작은 수로 나눈 나머지를 구하라
- 큰 수를 작은 수로 대체하고 작은 수는 나머지로 대체하라
- 작은 수가 0이 될 때까지 이 과정을 반복하라. 마지막 큰 수가 최대 공약수다.

```python
a, b = map(int, input().split(','))
big = max(a,b)
small = min(a,b)

while small != 0:
    rest = big % small
    big = small
    small = rest

print(big)
```

### 숫자를 문자열로 변환하는 방법은 str(num)을 이용한다. str(12) -> '12'가 된다.
### 반대로 문자열을 숫자로 변환하려면 int(string)을 이용한다.  int('12') -> 12가 된다.
### 이를 이용하여 1부터 1000까지의 숫자를 각 자리수의 합을 모두 구하라.
### 예를 들어 234 -> 2+3+4=9가 된다.
- sum = 0
for s in '234':
    sum += int(s)

```python
for i in range(1000):
    sum = 0
    for j in str(i):
        sum += int(j)
    print(i, sum)
```

### 다음 Person 클래스를 상속 받는 Employee 클래스를 정의하라. 
### Employee 클래스에 employeeID 속성을 추가하고 getID() 메소드를 정의하라.
### getID() 메소드는 employeeID를 반환하는 메소드이다.
### 최종적으로 Employee 클래스를 이용하여 Employee("IoT", 65, 2018)로 생성된 객체의 이름, 나이, ID를 출력하는 프로그램을 작성하여라.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def getName(self):
        print(self.name)
        
    def getAge(self):
        print(self.age)
        
        
class Employee(Person):
    def __init__(self, name, age, ID):
        super().__init__(name, age)
        self.ID = ID
        
    def getID(self):
        print(self.ID)

person = Employee("IoT",64,2018)
person.getName()
person.getAge()
person.getID()
```
